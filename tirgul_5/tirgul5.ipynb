## Let's plot some practicle plots
Read the *trees.csv* file and display height vs volume graph

```python
import pandas as pd
url = 'trees.csv'
treesData = pd.read_csv(url)
treesData.head()
```

```python
treesHeight = treesData['Height']
treesVolume = treesData['Volume']
```

```python
plt.plot(treesHeight,treesVolume,'ro') # b- is for 'bule', 'o' sets the marker shape to be a circle. For more colors/shapes look a the docs
plt.title('Trees Volume vs Height')
plt.ylabel('Volume')
plt.xlabel('Height')
```

### Display the *boxplot*
+ O is an outlier
+ \- represents the max/min values that are not outliers
+ red line is the median
+ The box top/bottom represent the 2nd and 3rd quarters of the data

```python
myData = [treesData['Height'] , treesData['Volume']]
bp=plt.boxplot(myData)
```

### Get the data from the boxplot

```python
bp
```

```python
bp['whiskers'][0].get_data()
```

### Histogram:

```python
plt.hist(treesData['Height'] )
plt.ylabel('# num of appearances')
plt.xlabel('height')
plt.title('Histogram of trees height')
```

```python
import numpy as np
import matplotlib.pyplot as plt
res = 10
x = np.linspace(0,10*np.pi , res)
print("X: {}...{}".format(x[:5],x[-5:]))
y = np.sin(x)
print("Y: {}...{}".format(y[:5],y[-5:]))
plt.plot(x,y)
```

### Subplots

```python
# We can plot several subplots in one figure
fig = plt.figure() # create a plot figure
# create the first of two panels and set current axis
import numpy as np
x = np.linspace(-2*np.pi, 2*np.pi, 100)
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))
plt.title('sin x')
# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x), color = 'red' , linestyle = '--'); # --r will also work
plt.title('cos x',color = 'green', fontsize = 20)
```

### Titles

```python
x = np.linspace(-10,10,1000)
plt.plot(x,x**2,'--y')
plt.xlabel('x')
plt.ylabel('$y=x^2$')
plt.title('$y=x^2$')
```

### Plot X/Y axis limits

```python
# we can control the axis limits:
plt.plot(x, np.sin(x))
plt.xlim(-4,4)
plt.ylim(-2,2)
plt.show()
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight');
```

```python
# Set equal ratio
plt.plot(x, np.sin(x))
plt.axis('equal');
```

### Plot Legend

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':r', label='cos(x)')
plt.legend();
```
