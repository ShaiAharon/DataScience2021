```python
import numpy as np
lst = [1, 2, 3, 4, 5]
arr = np.array([
    [1,2,3,4],
    [11,22,33,44]
])
arr.shape
arr1 = np.array([
    [11,21,31,1],
    [11,21,31,1]
])
arr * arr1
# print(arr*arr)
# arr.dot(arr.T)
```

```python
#### vector multiplication (AKA dot producdt)
print("arr.dot(arr)\t=",arr.dot(arr1.T))
# # or by 
print("arr@arr\t\t=",arr @ arr1.T)
```

```python
import numpy as np
x = np.array([1, 2, 3, 4])
print("Original array:")
print(x)
print("Test if none of the elements of the said array is zero:")
print(np.all(x))
x = np.array([0, 1, 2, 3])
print("Original array:")
print(x)
print("Test if none of the elements of the said array is zero:")
print(np.all(x))
```
